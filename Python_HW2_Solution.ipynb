```python
# JTMS-14
# a2_p5.py
# Roshan Shah
# roshah@jacobs-university.de

import numpy as np 
import math
import scipy


f1 = float(input()) 
f2 = float(input())

i1 = int(input())
i2 = int(input())


# Float Operations

sum_f = f1 + f2
diff_f = f2 - f1
sqr_f = math.pow(f2,2)

# Integer Operations

sum_i = i1 + i2
prod_i = i1 * i2

int_div = i1 // i2
real_div = i1 / i2
rem = i1 % i2

pow_i = int(math.pow(i1,i2))

print('sum =',sum_f)
print('diff =',diff_f)
print('square =',sqr_f)
print('sum =',sum_i)
print('prod =',prod_i)
print('int. div =',int_div)
print('div =',real_div)
print('remainder =',rem)
print('pow =',pow_i)


```
