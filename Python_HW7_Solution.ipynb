```python
# JTMS-14
# a7_p8.py
# Roshan Shah
# roshah@jacobs-university.de

stack = []


des = False

while not des:

  lett = input()

 
  if (lett == 's'):
    num = int(input())

    stack.append(num)
    print("Pushing %i" %num)
 
  elif (lett == 'p'):
    if(stack != []):
      print("Popping element %i" %stack.pop())

    else:  
      print('Stack underflow')

  elif (lett == 'e'):
    print("Emptying Stack: ")

    for i in range(len(stack)):
      print("Popping element %i" %stack.pop())
    
  elif (lett == 'q'):
    print("Good Bye !")
    des = True
 


```
