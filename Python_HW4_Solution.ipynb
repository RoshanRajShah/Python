```python
# JTMS-14
# a4_p6.py
# Roshan Shah
# roshah@jacobs-university.de

def print_frame(n,m,c):

  str1 = ''

  for i in range(0,n):

    if (i == 0 or i == n-1):

      for j in range(0,m): 
        str1 = str1 + c 
      print(str1)
      str1 = ''
    
    else:
      for j in range(0,m):
        if (j == 0 or j == m-1):
          str1 = str1 + c
        else:
          str1 = str1 + ' '
      print(str1)
      str1 = ''



n,m,c = int(input()), int(input()), input() 

print_frame(n,m,c)
```
