```python
# JTMS-14
# a5_p4.py
# Roshan Shah
# roshah@jacobs-university.de

import math 
import numpy as np
import random

random.seed()

minval = 1
maxval = 50

r = random.randint(minval,maxval)

found = True
count = 0

while found :

  guess = int(input("Enter your guess : "))

  if r == guess :

    print ( " You got it ! " )
    break

  elif guess < r :
    count = count +1 
    print ( " Your guess is too small ! , Attempt Count = %i" %count )

  else :
    count = count +1 
    print ( " Your guess is too high ! , Attempt Count = %i" %count )

print ('Total Attempt = %i' %count)
print ( " Bye " )


```

```python
# JTMS-14
# a5_p6.py
# Roshan Shah
# roshah@jacobs-university.de

import math 
import numpy as np
import random


def count_vowels(s):

  arr1 = list(s)
  n1 = len(arr1)
  
  arr2 = list(['a','e','i','o','u','A','E','I','O','U'])
  n2 = len(arr2)

  count = 0

  for i in range(n1):
    for j in range(n2):
      if (arr1[i] == arr2[j]):
        count += 1


  return count


op = False

while not op:

  s = input()

  if(len(s) == 0):
    op = True
    #print('No string entered.')
    break

  else:
    k = count_vowels(s)
    print('Number of vowels = ',k)
  
```
